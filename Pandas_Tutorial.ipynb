#Aggregate Statistics (Groupby)

```python
df = pd.read_csv('/content/modified (1).csv')

df
```

```python
df = pd.read_csv('/content/modified (1).csv')

num_cols = ['Speed','Generation','Total','HP','Attack','Defense','Sp. Atk','Sp. Def']

df.groupby('Type 1')[num_cols].mean()


```

```python
num_cols = ['Speed','Generation','Total','HP','Attack','Defense','Sp. Atk','Sp. Def']

df.groupby('Type 1')[num_cols].mean().sort_values('Defense', ascending=False)
```

```python
num_cols = ['Speed','Generation','Total','HP','Attack','Defense','Sp. Atk','Sp. Def']

df.groupby('Type 1')[num_cols].mean().sort_values('HP', ascending=False)
```

```python
num_cols = ['Speed','Generation','Total','HP','Attack','Defense','Sp. Atk','Sp. Def']

df.groupby('Type 1')[num_cols].sum()
```

```python
num_cols = ['Speed','Generation','Total','HP','Attack','Defense','Sp. Atk','Sp. Def']

df.groupby('Type 1')[num_cols].count()
```

```python
num_cols = ['Speed','Generation','Total','HP','Attack','Defense','Sp. Atk','Sp. Def']

df['count'] = 1

df.groupby('Type 1')[num_cols].count()
```

#Working with Large Amount Of Data

```python
for df in pd.read_csv('/content/modified (1).csv', chunksize=5):
  print("CHUNK DF")
  print(df)
```

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('/content/modified (1).csv', chunksize=5):
  results = df.groupby(['Type 1']).count()

  new_df = pd.concat([new_df, results])

  print(new_df)
```
